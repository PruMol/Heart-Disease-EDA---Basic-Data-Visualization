# Heart Disease - Exploratory Data Analysis


Heart disease remains a significant global health concern, affecting millions of individuals worldwide. Exploratory Data Analysis (EDA) is a vital approach used by researchers and analysts to gain insights and understand patterns within datasets related to heart disease. 

During EDA, various statistical techniques and visualizations are employed to explore the data's structure and distribution. These may include summary statistics, histograms, scatter plots, box plots, correlation matrices, and more. By analyzing these representations, analysts can identify trends, dependencies, and potential data issues, such as missing values or outliers.




OBJECTIVE: We shall delve into the heart disease data, analyze its key characteristics, and unravel valuable insights that may aid in prevention, diagnosis, and treatment strategies.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('heart.csv')
```

```python
df.head()
```

## Distribution Plot
#### In seaborn, displot is a function that plots histogram, kernel density estimate (KDE) plot and/or rug plot on a single variable. It is used to visualize the distribution of a single continuous variable.

```python
# Distribution plot (Displot) on Age column
sns.displot(df['Age'],kde=True)
```

```python
# Distribution plot (Displot) on Resting BP
sns.displot(df['Sex'],color='orange')
```

INFERENCE1: After doing the visualizations, we can infer that minimum age is around 30 and maximum age is around 80. Maximum of data falls within the range of 40 to 70. Male count is far more greater than female count.

## Pie charts for categorical columns
#### A piechart is a circular statistical graph, which is divided into slices to illustrate numerical proportions. Pie charts are generally used to show the distribution of a categorical variable.


```python
# Sex
df.groupby('Sex').size().plot(kind='pie',autopct='%0.1f', figsize=(4,4))
```

```python
# Chest Pain Type
df.groupby('ChestPainType').size().plot(kind='pie',autopct='%0.1f',figsize=(4,4))
```

INFERENCE2: 79% of males and 21% of females are having heart diseases.

## Violin Plot
#### In seaborn, a violinplot is a graphical representation of a continuous distribution, showing the probability density data at different values.


```python
sns.violinplot(y=df['Sex'],x=df['HeartDisease'])
```

```python
sns.violinplot(y=df['Age'],x=df['HeartDisease'])
```

INFERENCE3: Males are having more chances of having a heart disease than females. Whereas people belonging to the age group 58-60 have high chances of having a heart disease, as they lie higher than the mean (midpoint).

## Heatmap
#### A correlation heatmap is a graphical representation of the correlation matrix of dataset, which shows the the correlation coefficients (Pearson's correlation coefficients) between different variables in a dataset. The colors in the heatmap represent the strength of correlation with darker colors indicating a stronger correlation. '1' indicates a strong positive correlation, '-1' indicates a strong negative correlation and '0' indicates no correlation.

```python
# Converting strings to float (numeric values)
df['Sex'] = pd.to_numeric(df['Sex'], errors='coerce')
df['ChestPainType'] = pd.to_numeric(df['ChestPainType'], errors='coerce')
df['RestingECG'] = pd.to_numeric(df['RestingECG'], errors='coerce')
df['ExerciseAngina'] = pd.to_numeric(df['ExerciseAngina'], errors='coerce')
df['ST_Slope'] = pd.to_numeric(df['ST_Slope'], errors='coerce')
```

```python
sns.heatmap(df.corr())
```

## Jointplot
#### A correlation jointplot is a graphical representation of joint distribution of two variables, along with their individual distributions. The plot includes a scatterplot of data, with a regression line showing the trend, as well as the histograms or kernel density plots of individual variables on the X and Y axes.

```python
sns.jointplot(x='Age',y='MaxHR',data=df, kind='reg')
```

INFERENCE3: By the jointplot, we can infer that MaxHR decreases as the age increases.

## PairPlot
#### A correlation pairplot is a graphical representation of the pairwise relationships between variables in a dataset. It is a matrix of scatter plots, where each scatter plot shows the relationship between two variables. The diagonal of the matrix is a histogram or kernel density plot showing the distribution of each variable (one of the variables).

```python
sns.pairplot(df)
```
